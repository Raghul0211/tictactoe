```python
mat=[1,2,3,4,5,6,7,8,9]
status="running"
i=0
def display():
        print("{}  | {} | {}".format(mat[0],mat[1],mat[2]))
        print("___|___|___")
        print("{}  | {} | {}".format(mat[3],mat[4],mat[5]))
        print("___|___|___")
        print("{}  | {} | {}".format(mat[6],mat[7],mat[8]))
        print("   |   |  ")
def check(ck):
    if(ck>=10 or ck<1 or mat[ck-1]=="X" or mat[ck-1]=="O"):
        chk=eval(input("You cannot mark here try different place"))
        new=check(chk)
    else:
        new=ck
    return new
def update(val,symb):
    mat[val-1]=symb
def winorloss():
    global status
    if ((mat[0]==mat[1] and mat[1]==mat[2])or(mat[0]==mat[3] and mat[3]==mat[6])):
        if mat[0]=="X":
            print("player 1 won")
            status="winner"
        else:
            print("player 2 won")
            status="winner"
        display()
    if ((mat[3]==mat[4] and mat[4]==mat[5])or(mat[1]==mat[4] and mat[4]==mat[7])):
        if mat[4]=="X":
            print("player 1 won")
            status="winner"
        else:
            print("player 2 won")
            status="winner"
        display()
    if ((mat[6]==mat[7] and mat[7]==mat[8])or(mat[2]==mat[5] and mat[5]==mat[8])):
        if mat[8]=="X":
            print("player 1 won")
            status="winner"
        else:
            print("player 2 won")
            status="winner"
        display()
    if ((mat[0]==mat[4] and mat[4]==mat[8])or(mat[2]==mat[4] and mat[4]==mat[6])):
        if mat[4]=="X":
            print("player 1 won")
            status="winner"
        else:
            print("player 2 won")
            status="winner"
        display()
while status=="running":
    display()
    val1=eval(input('player 1:which location you want to put "X"'))
    p1=check(val1)
    update(p1,"X")
    winorloss()
    i=i+1
    if(i==9 and status=="running"):
        display()
        print("game draw")
        break
    if status=="winner":
        break
    display()
    val2=eval(input('player 2:which location you want to put "O"'))
    p2=check(val2)
    update(p2,"O")
    winorloss()
    i=i+1
    
    
```
